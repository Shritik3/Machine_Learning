```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# epsilon and minimum no of points
# it focuses on density more than distance which is used in kmeans
```

```python
# Theree main point types
# Core,border,outlier
# Randomly choose point
# Epsilon is the distance from the random point to identify its relationship with other points
# min no points within an epsilon neighborhood to decide if it is a core point or not
# Core point - point with min pointsin epsilon range
# border point-epsilon range of core point but not have min no of points(including or excluding the point)
# if a point can't be reached in a cluster assignment then it is a outlier which it identifies on its own
```

```python
# epsilon large points will come in same cluster 
# very unique then it will have too many no of clusters
```

```python
# Best choice for hyper-parameters
# search for best value of parameters 
# elbow or knee method with varying epsilon value
# prior knowledge of clusters and know the relative percentage of outliers
# make a plot of min no of points with outliers
```

```python
two_blobs=pd.read_csv('../DATA/cluster_two_blobs_outliers.csv')
```

```python
sns.scatterplot(data=two_blobs,x='X1',y='X2')
```

```python
def display_categories(model,data):
    labels=model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db=DBSCAN()
```

```python
display_categories(db,two_blobs)
```

```python
db=DBSCAN(eps=0.001)
display_categories(db,two_blobs)
# everything is outlier
```

```python
db=DBSCAN(eps=10)
display_categories(db,two_blobs)
# all fitting under same cluster
```

```python
db=DBSCAN(eps=1)
display_categories(db,two_blobs)
```

```python
db.labels_
```

```python
np.sum(db.labels_==-1)
# no of outliers
```

```python
100*np.sum(db.labels_==-1)/len(db.labels_)
```

```python
np.unique(db.labels_)
```

```python
# elbow plot to see dramactic change by tuning hyperparameter
outlier_percent=[]
num_of_outlier=[]
for eps in np.linspace(0.001,7,200):
    db=DBSCAN(eps=eps)
    db.fit(two_blobs)
    num_of_outlier.append(np.sum(db.labels_==-1))
    outlier_perc=100*np.sum(db.labels_==-1)/len(db.labels_)
    outlier_percent.append(outlier_perc)
```

```python
sns.lineplot(x=np.linspace(0.001,7,200),y=num_of_outlier)
# it tells 
```

```python
sns.lineplot(x=np.linspace(0.001,7,200),y=num_of_outlier)
plt.xlim(0,2)
```

```python
sns.lineplot(x=np.linspace(0.001,7,200),y=outlier_percent)
plt.xlim(0,2)
```

```python
sns.lineplot(x=np.linspace(0.001,7,200),y=outlier_percent)
plt.xlim(0,2)
plt.ylim(0,10)
plt.hlines(y=3,xmin=0,xmax=2,color='red')
# to get three outliers
```

```python
sns.lineplot(x=np.linspace(0.001,7,200),y=num_of_outlier)
plt.xlim(0,2)
plt.ylim(0,10)
plt.hlines(y=3,xmin=0,xmax=2,color='red')
# poits till 1 only vary
```

```python
outlier_percent=[]
num_of_outlier=[]
for n in np.arange(1,100):
    db=DBSCAN(min_samples=n)
    db.fit(two_blobs)
    num_of_outlier.append(np.sum(db.labels_==-1))
    outlier_perc=100*np.sum(db.labels_==-1)/len(db.labels_)
    outlier_percent.append(outlier_perc)
```

```python
# min_samples
sns.lineplot(x=np.arange(1,100),y=outlier_percent)
```

```python
num_dim=two_blobs.shape[1]#for default start
db=DBSCAN(min_samples=2*num_dim)
display_categories(db,two_blobs)
```

```python
db=DBSCAN(min_samples=1,eps=0.75)
display_categories(db,two_blobs)
```
