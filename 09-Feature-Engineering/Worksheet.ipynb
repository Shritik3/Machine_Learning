```python
# Range and limits for outlier
# Range=1.Interquartile range(Can be changed according to data) 2.Standard deviation of dat
# Percentage of data-Few percentage of data for outliers
# Visualisation(Scatter plot)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Choose a mean,standard deviation, and number of samples

def create_ages(mu=50,sigma=13,num_samples=100,seed=42):

    # Set a random seed in the same cell as the random call to get the same values as us
    # We set seed to 42 (42 is an arbitrary choice from Hitchhiker's Guide to the Galaxy)
    np.random.seed(seed)

    sample_ages = np.random.normal(loc=mu,scale=sigma,size=num_samples)
    sample_ages = np.round(sample_ages,decimals=0)
    
    return sample_ages
```

```python
sample=create_ages()
```

```python
sample
```

```python
# Check distribution
sns.displot(data=sample,bins=20)
```

```python
sns.boxplot(x=sample)
```

```python
series=pd.Series(sample)
```

```python
# IQR is between 25 to 75%
series.describe()
```

```python
IQR=55.25000-42.00000
lower_limit=42.0-1.5*(IQR)
```

```python
lower_limit
```

```python
# To get rid of the outlier
series[series>lower_limit]
```

```python
# Another way to get IQR
q75,q25=np.percentile(sample,[75,25])
```

```python
def percent_missing(fd):
    percent_nan=100*fd.isnull().sum()/len(fd)
    percent_nan=percent_nan[percent_nan>0].sort_values()
    
    return percent_nan
```
