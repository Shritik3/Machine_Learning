```python
# assigns label to new data on the distance between old and new data
# right number of points to consider for calculating distance from new point
# k=1 means find the closest point to the new upcoming point and and decide it to the same label
# k=2 check for two closest points if both belong to same then assign it to that class
# k=3 if two are close to one class and one is close to other class then we'll take the majority and decide the class of the new datapoint
# in k=4 if there is a tie between two classes then:
# 1.always choose odd k
# 2.reduce k by 1 until tie is broken
# 3.randomly choose any class
# 4. or then go to k=1 and choose the class same as the nearest class
# ordering of data matters class which appear first will be chosen
# find optimal value of k
# error=1-accuracy
# grid search for lowest k and highest accuracy
# Elbow method (error vs k)
# as k increases (complexity )but the error is reducing by very small percentage
# Cross validation with grid search 
# choose k
# sort features by distance
# choose class based on k nearest feature
# distance manhattan,euclidean,minkowski
# scale data is necessary for the distance metric
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=df.drop('Cancer Present',axis=1)
```

```python
y=df['Cancer Present']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
help(KNeighborsClassifier)
```

```python
knn_model=KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(X_train,y_train)
```

```python
y_pred=knn_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
len(y_test)
```

```python
print(classification_report(y_test,y_pred))
# precision and recall are in line with accuracy(no imbalanced class set)
```

```python
df['Cancer Present'].value_counts()
```

```python
# Ideal k value
# 1.Gridsearch cross validation or elbow method
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
error=1-accuracy_score(y_test,y_pred)
error
```

```python
# sequence of repeated operation by pipeline object like calling scaller and model
test_error_rates=[]

for k in range(1,30):
    knn_model=KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train)
    y_pred_test=knn_model.predict(X_test)
    test_error=1-accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
```

```python
# low k value high error rate
test_error_rates
# after too high k see for leveling off or a spike
```

```python
plt.plot(range(1,30),test_error_rates)
plt.ylabel('ERROR RATES')
plt.xlabel('K Neighbors')
# Exact minimum of k is  around 22
# But we to trade off between complexity and amout of error reduced that is till k=6
# Depends on context of data
```

```python
# PIPELINE-->Gridsearch cv
```

```python
scaler=StandardScaler()
```

```python
knn=KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python
operations=[('scaler',scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# CV for all k values
k_values=list(range(1,20))
```

```python
k_values
```

```python
# param_grid={'n_neighbors':k_values} can't use this with the pipeline
# We need to use the string name like this
param_grid={'knn__n_neighbors':k_values}
```

```python
full_cv_classifier=GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
```

```python
# no need to do scale as pipe is used
full_cv_classifier.fit(X_train,y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
full_pred=full_cv_classifier.predict(X_test)
```

```python
# 'knn__n_neighbors': 18
print(classification_report(y_test,full_pred))
```
