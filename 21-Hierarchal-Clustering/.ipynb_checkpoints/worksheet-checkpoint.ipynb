```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_d=pd.get_dummies(df.drop('name',axis=1))
```

```python
df_d
```

```python
from sklearn.preprocessing import MinMaxScaler
# to get max distance between points
```

```python
sc=MinMaxScaler()
```

```python
sc_df=sc.fit_transform(df_d)
```

```python
sc_df
# max range between 0 to 1
```

```python
sc_df=pd.DataFrame(sc_df,columns=df_d.columns)
```

```python
sc_df
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(sc_df,cmap='viridis')
```

```python
# number of cars
len(df)
```

```python
# model year has nice gradient as it is in sequence
# random color because of continuous data
# and origin is dummy variable that's why only two colour of 0 and 1
plt.figure(figsize=(15,8))
sns.clustermap(sc_df,cmap='viridis')
# cluster similar cars in group and not columns
```

```python
plt.figure(figsize=(15,8))
sns.clustermap(sc_df,cmap='viridis',row_cluster=False)
# dummy variable grouped last as they have max dist between them i.e, 0 and 1
```

```python
# we want this
plt.figure(figsize=(15,8))
sns.clustermap(sc_df,cmap='viridis',col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=4)
```

```python
sc_df
```

```python
cluster_labels=model.fit_predict(sc_df)
```

```python
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels)
```

```python
sns.scatterplot(data=df,x='mpg',y='horsepower',hue=cluster_labels,
               palette='viridis')
```

```python
# choose the numer of cluster
len(sc_df.columns)
```

```python
np.sqrt(10)#max dist possible between two points(Min max scaling)
```

```python
model=AgglomerativeClustering(n_clusters=None,distance_threshold=0)
# linkage distance threshold above which clusters will not be merged
```

```python
cluster_labels=model.fit_predict(sc_df)
```

```python
cluster_labels
# unique to every point
# max no of clusters possible
```

```python
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram
```

```python
linkage_metrix=hierarchy.linkage(model.children_)
```

```python
linkage_metrix
#  67.        , 161.        ,   1.41421356,   2.
# 67 and 171 two points and 1.41 is distance
# 349.        , 779.        ,  75.32595834, 391.   
# the top linkage and the number of points connected under it
```

```python
plt.figure(figsize=(20,10),dpi=300)
dendro=dendrogram(linkage_metrix)
```

```python
plt.figure(figsize=(20,10),dpi=300)
dendro=dendrogram(linkage_metrix,truncate_mode='lastp',p=10)
```

```python
plt.figure(figsize=(20,10),dpi=300)
dendro=dendrogram(linkage_metrix,truncate_mode='level',p=3)
```
