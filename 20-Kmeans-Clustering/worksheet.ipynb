```python
X=pd.get_dummies(df)
```

```python
X
```

```python
# scale as we are dealing with distance metrics
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
# No test st and y label
scaled_X=scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
help(KMeans)
```

```python
# more clusters more max iterations
model=KMeans(n_clusters=2,random_state=101)
# random number of clusters(number of people market to and not market to)
```

```python
cluster_label=model.fit_predict(scaled_X)
# fitting and returning cluster labels
```

```python
cluster_label
```

```python
X['Cluster']=cluster_label
```

```python
X
```

```python
plt.figure(figsize=(12,6),dpi=300)
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
plt.xticks(rotation=90);
```

```python
# sum of squared distance from center to punish points which are far from cluster center
# We check SSD for different k as it should always go dows as k increases 
# it can be perfect at clusters equals to 0 we need to check rate of reduction and check the trade off
# using elbow method
```

```python
ssd=[]
for k in range(2,20):
    model=KMeans(n_clusters=k)
    model.fit(scaled_X)
    ssd.append(model.inertia_)# SSD point --.cluster center
```

```python
ssd
```

```python
plt.plot(range(2,20),ssd,'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

### Colour Quantization

```python
# for coloured image
# (1280,720,3)
# width,height,color channels(red,green blue)
# simplify the number of colours or decrease the quality of image 
# the three array coordites have different value for pixels
# reshape the 3d array and flatten it H*W every row to a datapoint
# then use k means clustering
# cluster center is the avh of rgb in a range of values
# reducing colour space to  number of colours--> color quantization
```

```python
import matplotlib.image as mpimg
# image processing
```

```python
# for reading jpg or png-->numpy array
image_as_array=mpimg.imread('../DATA/palm_trees.jpg')
```

```python
image_as_array
```

```python
image_as_array.shape#(h,w,c)
```

```python
plt.figure(dpi=500)
plt.imshow(image_as_array)
# imshow needs numpy array
```

```python
# kmeans to quantize colours
# (h,w,c)3D--.2d(h*w,c)
(h,w,c)=image_as_array.shape
```

```python
c
```

```python
h
```

```python
image_as_array2d=image_as_array.reshape(h*w,c)
```

```python
image_as_array2d
```

```python
len(image_as_array.shape)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=6)
```

```python
# k means need array in 2d
labels=model.fit_predict(image_as_array2d)
```

```python
labels
# 1st three and last three should be same
```

```python
rgb_codes=model.cluster_centers_.round(0).astype(int)
# six avg out rgb values
```

```python
rgb_codes
```

```python
# labels are associated with rgb_codes
# 3 cluster matches to  [ 71, 109, 138] this rgb code
rgb_codes[labels]
# reassign points to cluster centers of those minimalistic colours
```

```python
image_as_array2d
```

```python
quantized_image=np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantized_image
```

```python
image_as_array
```

```python
plt.figure(dpi=500)
plt.imshow(quantized_image)
# only 6 colours as k was 6
```
