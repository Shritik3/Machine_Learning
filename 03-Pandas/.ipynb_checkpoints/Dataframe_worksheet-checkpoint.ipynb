```python
def quality(total_bill,tip):
    if tip/total_bill>0.25:
        return "Generous"
    else:
        return "Other"
```

```python
quality(16.99,.101)
```

```python
df['quality']=df[['total_bill','tip']].apply(lambda df:quality(df['total_bill'],df['tip']),axis=1)
```

### Merge

```python
# Inner merge looking for matches on both table and to select a column which is also unique
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
help(pd.merge)
```

```python
pd.merge(registrations,logins,how='inner',on='name')
# order of tables doesn't matter
```

```python
# Left join or right join
```

```python
# how='left'
# regis on left and logins on right
# It means all the info present in regis table will be displayed regardless if it is common or not in the other table
# Null values will come which are not common in both the tables
pd.merge(registrations,logins,'left','name')
```

```python
# how='right'
# Order matters for left and right merge
pd.merge(registrations,logins,'right','name')
```

```python
# Outer join for grabbing all the info of both the tables
pd.merge(registrations,logins,'outer','name')
```

```python
registrations=registrations.set_index('name')
```

```python
registrations
```

```python
# left_on to add a column
# left_index to have the index of the table
pd.merge(registrations,logins,left_index=True,right_on='name',how='inner')
```

```python
registrations=registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns=['reg_name','reg_id']
```

```python
registrations
```

```python
logins
```

```python
res=pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name')
```

```python
res.drop('reg_name',axis=1)
```

```python
registrations.columns=['name','id']
```

```python
logins.columns=['id','name']
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on='name')
# automatically deletes the duplicate id value and automatically gives new default suffixes
```

```python
pd.merge(registrations,logins,how='inner',on='name',suffixes=('_reg','_log'))
```
