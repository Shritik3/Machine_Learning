```python
a=np.linspace(0,10,11)
```

```python
b=a**4
```

```python
a
```

```python
b
```

```python
x=np.arange(0,10)
```

```python
y=2*x
```

```python
x
```

```python
y
```

```python
# Zoomed in a particular section of plot

fig=plt.figure()
# LARGE axis
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b);

axes.set_xlim(0,8)
axes.set_ylim(0,8000)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Power of 4')

# Small axis
axes1=fig.add_axes([0.2,0.4,0.45,0.45])
# axes1=fig.add_axes([1,1,0.45,0.45])
axes1.plot(a,b);
axes1.set_xlim(1,2)
axes1.set_ylim(0,100)
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('Zoom in power of 4')
```

```python
type(fig)
```

```python
# dpi for high resolution of graph
fig=plt.figure(figsize=(2,2),dpi=200)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('new_fig.png',bbox_inches='tight')
```

### Subplots

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
type(axes)
```

```python
axes.shape
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(2,2),dpi=300)
axes[0,0].plot(x,y)
axes[0][1].plot(x,y)
axes[0][1].set_title('Graph 2')
axes[1,1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][0].set_xlim(2,8)
# plt.tight_layout()

fig.suptitle('Figure Level',fontsize=16)
fig.set_figwidth(10)
fig.subplots_adjust(wspace=0.9,hspace=0.4)
fig.savefig('new_subplt.png',bbox_inches='tight')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
# axes[0].plot(a,b)
for fx in axes:
    fx.plot(x,y)
plt.tight_layout()
```

```python
axes.shape
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='X vs X')
ax.plot(x,x**2,label='Exponential of 2')
# ax.legend(loc='upper left')
# ax.legend(loc=(-0.5,0.5))
ax.legend(loc=(1.1,0.5))
```

### Styling

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,color='#650dbd',lw=5,ls='--',label='X vs X')
ax.plot(x,x+6,color='green',lw=2,ls='--',marker='s',markerfacecolor='red',markeredgewidth=1,markeredgecolor='black',label='X vs X')
ax.plot(x,x-2,color='yellow',lw=1,marker='o',ls='--',label='X vs X-2')
# Custom line style
# Default '--','-','-.',':','steps'
# matplotlib marker api
lines=ax.plot(x,x+4,color='red',linewidth=1.5,marker='+',ms=15,label='X vs X+4')
# ms or markersize
lines[0].set_dashes([5,2,10,2])
# [5,2,10,2] 5 len dash, 2 empty space, 10 line dash, 2 empty space 
ax.legend()
```
