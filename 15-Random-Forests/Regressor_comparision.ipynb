```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../DATA/rock_density_xray.csv')
```

```python
df.head()
```

```python
df.columns=['Signal','Density']
```

```python
sns.scatterplot(x='Signal',y='Density',data=df)
```

```python
# no scaling required as data is not varying much
# as have only one feature we have to reshape it to make sure it is a full data frame
X=df['Signal'].values.reshape(-1,1)
y=df['Density']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model=LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_pred=lr_model.predict(X_test)
```

```python
lr_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,lr_pred)
# it just provides the avg
```

```python
np.sqrt(mean_squared_error(y_test,lr_pred))
# these both error looks good and not extreme and avg performance
```

```python
signal_range=np.arange(0,100)
```

```python
signal_range
```

```python
signal_pred=lr_model.predict(signal_range.reshape(-1,1))
```

```python
signal_pred
# all are coming around 2.2 which is strange
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Signal',y='Density',data=df)

plt.plot(signal_range,signal_pred)
# it clearly shows a llinearly regression model cannot fit to this data
```

### Polynomial Regression

```python
def run_model(model,X_train,y_train,X_test,y_test):
    model.fit(X_train,y_train)
    
    preds=model.predict(X_test)
    rmse=np.sqrt(mean_squared_error(y_test,preds))
    mae=mean_absolute_error(y_test,preds)
    print(f'RMSE:{rmse}')
    print(f'MAE:{mae}')
    
    signal_range=np.arange(0,100)
    signal_preds=model.predict(signal_range.reshape(-1,1))
    
    plt.figure(figsize=(12,8))
    sns.scatterplot(x='Signal',y='Density',data=df,color='black')

    plt.plot(signal_range,signal_preds)
```

```python
model=LinearRegression()
run_model(model,X_train,y_train,X_test,y_test)
```

```python
# Create pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe=make_pipeline(PolynomialFeatures(degree=6),LinearRegression())
```

```python
run_model(pipe,X_train,y_train,X_test,y_test)
# more noise when we increase degree
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# too high k it starts flattening at 0 and 100
k_values=[1,5,10,20,35]
for n in k_values:
    model=KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svr=SVR()
param_grid={'C':[0.01,0.1,1,5,10,100,1000],'gamma':['auto','scale']}



grid=GridSearchCV(svr,param_grid)
```

```python
# lower variance model by knowledge of data
run_model(grid,X_train,y_train,X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(n_estimators=10)
```

```python
run_model(rfr,X_train,y_train,X_test,y_test)
# compare with polynomial results and choose the best 
# this fits best with noise and bias
```

```python
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
```

```python
model=GradientBoostingRegressor()
run_model(model,X_train,y_train,X_test,y_test)
# not much noise and out performing random forest
```

```python
model=AdaBoostRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```
